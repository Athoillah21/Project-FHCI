```python
df
```

```python
import numpy as np
def filtered_keywords(df, keywords, title="Keyword Analysis", head=15):
    # get keywords in a column
    count_keywords = pd.DataFrame(df.Jurusan_tokens.sum()).value_counts().rename_axis('keywords').reset_index(name='counts')

    # get frequency of occurence of word (as word only appears once per line)
    length = len(df) # number of job postings
    count_keywords['percentage'] = 100 * count_keywords.counts / length
    count_keywords['count'] = count_keywords.counts

    
    # plot the results
    count_keywords = count_keywords[count_keywords.keywords.isin(keywords)]
    count_keywords = count_keywords.head(head)
    # colors = sns.color_palette('pastel')[0:15]
    # colors = sns.color_palette('rocket')
    # colors = sns.color_palette('magma')
    # colors = sns.color_palette('deep')[0:15]
    # colors = sns.light_palette("seagreen")
    colors = sns.color_palette('crest', 15)
    
    g = plt.bar(x="keywords", height="count", data=count_keywords , color=colors)
    plt.xlabel("")
    plt.ylabel("Frekuensi Kemunculan")
    plt.xticks(rotation = 45, ha='right')
    plt.title(title)
    plt.show(g)

    print(count_keywords)
```

```python
filtered_keywords(df, keywords, title="Jurusan yang Sering Muncul dalam Lowongan")
```
