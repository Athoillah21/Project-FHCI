```python
import pandas as pd

df = pd.read_csv(f'/home/athoillah/Documents/FHCI/FHCI_Desember2022.csv')
```

```python
df = df.replace('\n',',', regex=True)
df = df.replace('Requirement:',' ', regex=True)
df = df[df.Posisi != 'Empty Cause Error Network']
```

```python
df['Induk Perusahaan'] = df['Perusahaan'].str.split(' - ').str[0]
```

```python
df.drop(['Jurusan', 'Requirement', 'Link'], axis=1, inplace=True)
```
